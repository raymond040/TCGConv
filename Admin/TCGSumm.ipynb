```python
import sys
import math
from contextlib import redirect_stdout


def calculate_turns(seats, customer, dishes):
    # Write your code here
    circular = dishes+dishes
    cnt_epd = 0
    turns = 0
    cur = len(dishes)+customer
    while cnt_epd <3:
        cur_dct = {}
        abs_dct = {}
        #find the nearest location
        for i in range(len(circular)):
            if circular[i] not in cur_dct:
                cur_dct[circular[i]] = [i-cur]
                abs_dct[circular[i]] = [abs(i-cur)]
            else:
                cur_dct[circular[i]].append(i-cur)
                abs_dct[circular[i]].append(abs(i-cur))
        if cnt_epd ==0:
            cur_turn = min(abs_dct['E'])
            turns += cur_turn
            cnt_epd += 1
            if cur_turn in cur_dct['E']:
                cur += min(abs_dct['E'])
            else:
                cur -= min(abs_dct['E'])
            continue
        elif cnt_epd ==1:
            cur_turn = min(abs_dct['P'])
            turns += cur_turn
            cnt_epd += 1
            if cur_turn in cur_dct['P']:
                cur += min(abs_dct['P'])
            else:
                cur -= min(abs_dct['P'])
            continue
        elif cnt_epd ==2:
            cur_turn = min(abs_dct['D'])
            turns += cur_turn
            cnt_epd += 1
            if cur_turn in cur_dct['D']:
                cur += min(abs_dct['D'])
            else:
                cur -= min(abs_dct['D'])
    return turns
        
calculate_turns(3,0,"EPD") #ENPDNNN ENPDNNNN
```
